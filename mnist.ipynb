```python

import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.optimizers import RMSprop
import os

num_classes = 10
def data_generator_mnist(isTrain = True, batchSize = 100):

    (X_train, y_train),(X_test, y_test) = mnist.load_data()
    X_train = X_train.reshape(60000,784)
    X_test = X_test.reshape(10000,784)
    X_train = X_train.astype('float32')
    X_test = X_test.astype('float32')
    X_train /= 255
    X_test /= 255
    
    Y_train = keras.utils.to_categorical(y_train, num_classes)
    Y_test = keras.utils.to_categorical(y_test, num_classes)
    
    if isTrain:
        dataset = (X_train, Y_train)
    else:
        dataset = (X_test, Y_test)
        
    dataset_size = dataset[0].shape[0]
    i = 0
    while 1:
        x,y = dataset[0][i:i+batchSize],dataset[1][i:i+batchSize]
        yield x,y
        i = (i+1) % (dataset_size// batchSize)
            
model = Sequential()
model.add(Dense(512,activation='relu',input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_classes,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer=RMSprop(),metrics=['accuracy'])


history = model.fit_generator(data_generator_mnist(True,batchSize=100),steps_per_epoch=600,epochs=20)
```

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.optimizers import RMSprop

batch_size = 256
num_classes = 10
epochs = 20

(x_train, y_train),(x_test, y_test) = mnist.load_data()

x_train = x_train.reshape(60000,784)
x_test = x_test.reshape(10000,784)

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

x_train /= 255
x_test /= 255

y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

model = Sequential()
model.add(Dense(512,activation='relu',input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_classes,activation='softmax'))

model.compile(loss='categorical_crossentropy',optimizer=RMSprop(),metrics=['accuracy'])

model.fit(x_train,y_train,batch_size = batch_size,epochs = epochs, verbose=1,validation_data=(x_test,y_test))
```
